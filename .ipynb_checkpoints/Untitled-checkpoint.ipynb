```python
# Feature Selection - Numerical Features
# Part 1 : Remove Features with 0 Variance
```

```python
from sklearn.feature_selection import VarianceThreshold

varselector= VarianceThreshold(threshold=0)
varselector.fit_transform(num_col)
# Get columns to keep and create new dataframe with those only
cols = varselector.get_support(indices=True)
num_1 = num_col.iloc[:,cols]
```

```python
num_1
```

```python
from sklearn.preprocessing import KBinsDiscretizer
discrete=KBinsDiscretizer(n_bins=10,encode='ordinal', strategy='quantile')
num_binned=pd.DataFrame(discrete.fit_transform(num_1),index=num_1.index, columns=num_1.columns).add_suffix('_Rank')
num_binned.head(10)
```

```python
X_bin_combined=pd.concat([Y,num_binned],axis=1,join='inner')

from numpy import mean
for col in (num_binned.columns):
    plt.figure()
    sns.barplot(x=col, y="target",data=X_bin_combined, estimator=mean )
plt.show()
```

```python
num_1.shape
```

```python
# Part 1 - Bi Variate Analysis¶
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
X_char_merged=pd.concat([Y,cat_col],axis=1,join='inner')

from numpy import mean
for col in (cat_col.columns):
    plt.figure()
    sns.barplot(x=col, y="target",data=X_char_merged, estimator=mean )
plt.show()
```

```python
# Create dummy features with n-1 levels
X_char_dum = pd.get_dummies(cat_col, drop_first = True)
X_char_dum.shape
```

```python
X_all=pd.concat([X_char_dum,num_binned],axis=1,join="inner")
```

```python
X_all.shape
```

```python
X_all.head()
```

```python
#Train test split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X_all, Y, test_size=0.3, random_state=10)
```

```python
#Balancing Dataset using SMOTE
```

```python
!pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
print("Shape of Training Data",X_train.shape)
print("Shape of Testing Data",X_test.shape)
print("Response Rate in Training Data",y_train.mean())
print("Response Rate in Testing Data",y_test.mean())
```

```python
# Shape of Training Data (28823, 53)
# Shape of Testing Data (12353, 53)
# Response Rate in Training Data 0.11487353849356417
# Response Rate in Testing Data 0.107504249979762
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(random_state=0)
logreg.fit(X_train_resampled,y_train_resampled)
```

```python
coeff_df=pd.DataFrame(X_all.columns)
coeff_df.columns=['features']
coeff_df["Coefficient Estimate"] = pd.Series(logreg.coef_[0])
coeff_df
```

```python
# Building a Decision Tree Model
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(criterion='gini',random_state=0)
```

```python
np.random.seed(44)
from sklearn.model_selection import GridSearchCV
param_dist = {'max_depth': [3, 5, 6, 7, 9], 'min_samples_split': [50, 80, 110, 130, 160, 190, 220, 250] }
tree_grid = GridSearchCV(dtree, cv = 10, param_grid=param_dist,n_jobs = 3)
tree_grid.fit(X_train_resampled,y_train_resampled) 
print('Best Parameters using grid search: \n', tree_grid.best_params_)
```

```python
dtree=DecisionTreeClassifier(criterion='gini',random_state=0,max_depth=5,min_samples_split=100)
dtree.fit(X_train,y_train)
```

```python
# Building a Random Forest Model
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(criterion='gini',random_state=0,max_depth=5,min_samples_split=100)
rf.fit(X_train_resampled,y_train_resampled)
```

```python
import pandas as pd
feature_importances=pd.DataFrame(rf.feature_importances_,
                                 index=X_train_resampled.columns,
                                 columns=['importance']).sort_values('importance',ascending=False)
feature_importances
```

```python
#XGBoost
from xgboost import XGBClassifier
```

```python
xgb_model = XGBClassifier(random_state=42)
xgb_model.fit(X_train_resampled, y_train_resampled)
```

```python
# Model Evaluation
y_pred_logreg=logreg.predict(X_test)
y_pred_tree=dtree.predict(X_test)
y_pred_rf=rf.predict(X_test)
y_pred_xgb = xgb_model.predict(X_test)
```
